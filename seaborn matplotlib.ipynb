```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[4,5,1],'^k')
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
x=[16,8,10]
y=[8,16,6]
x2=[5,6,7]
y2=[8,4,2]
plt.plot(x,y,'r',label='line one',linewidth=3)
plt.plot(x2,y2,'m',label='line two',linewidth=3)
plt.title("stylish graph")

plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
dept=['IT','HR','Network',"BPO"]
ren=[200000,1000000,400000,100000]
plt.barh(dept,ren,color='g')
plt.title("DEPT VS REVENUE")
plt.show()
```
